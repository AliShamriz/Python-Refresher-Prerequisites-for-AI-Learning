```python
import pandas as pd
```

```python
df = pd.read_csv(
    'Data/movies.csv',
    index_col='name',
    parse_dates=['release_date'],
)
df.head()
```

```python
above_7 = df['score']>=7
below_8 = df['score']<=8
df [above_7 & below_8]
```

```python
df['score'].between(7,8)
```

```python
df[df['score'].between(7,8)]
```

```python
df[df['release_date'].between('2000-01-01','2016-01-01')]
```

```python
df ['genre'].unique()
```

```python
is_mystery = (df['genre'] == 'Mystery')
is_scifi = (df['genre'] == 'Sci-Fi')
df [is_mystery | is_scifi]
```

```python
df[df ['genre'].isin(['Mystery','Sci-Fi'])]
```

```python
genres = ['Mystery','Sci-Fi','Adventure']
genres_filter = df['genre'].isin(genres)
df[genres_filter]
```
