```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4])
plt.ylabel('Funny Numbers')
plt.show()
```

```python
plt.plot([1,3,4,9],[-1,8,1,2])
```

```python
plt.plot([1,3,4,9],[-1,8,1,2],'ro')
```

```python
plt.plot([1,3,4,9],[-1,8,1,2],'bs')
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t)
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t**2)
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t*2)
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t**2,'r--')
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t**2,'r-.')
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t,'r--',t,t**2,'b.',t,t**3,'g^')
```

```python
data = {
    'a': np.arange(500),
    'b': np.random.randint(0,500,500),
    'c': np.random.randn(500)
}
```

```python
plt.scatter(data['a'],data['b'])
```

When using randn, the chances of the points being close to 0 are much higher.

```python
plt.scatter(data['a'],data['c'])
```
